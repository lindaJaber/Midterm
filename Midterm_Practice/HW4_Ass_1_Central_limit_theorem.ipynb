# Demonstrating the Central Limit Theorem 

```python
from __future__ import print_function

import numpy as np
import matplotlib as plt
import pylab as pl

%pylab inline
```

## Choosing the Distributions

Normal, Poisson, Binomial, Chi-Squared, and Laplace

```python
my_mean = 27
```

```python
np.random.seed(1234)
```

```python
my_size = np.random.choice(np.arange(11, 2000), 100, replace=False)
my_size, my_size.shape
```

```python
np.random.seed(1234)
my_size = np.random.choice((np.arange(11, 2000)), 100, replace=False)
print (np.array(sorted(my_size)), my_size.shape)
```

## Normal Distribution

```python
d={}
d['normal'] = {}

for int in my_size:
    d['normal'][int] = np.random.normal(my_mean, 1, int)   
```

```python
d['normal'].keys()
```

```python
print (np.array(sorted(d['normal'].keys())), len(d['normal']))
```

```python
d['normal']['means'] = {}

fig1 =  plt.figure(figsize=(12,6), dpi=100)
ax_normal = fig1.add_subplot(111)

for smpl in d['normal']:
    if not type(smpl) == str:
        d['normal']['means'][smpl] = d['normal'][smpl].mean()
        ax_normal.plot(smpl, d['normal']['means'][smpl], 'go')
        ax_normal.set_title('Normal Distribution', fontsize=20)
        ax_normal.set_xlabel('Sample Size', fontsize=15)
        ax_normal.set_ylabel('Sample Mean', fontsize=15)
        ax_normal.plot([min(my_size), max(my_size)], [my_mean, my_mean], 'm--')
```

**Figure 1:** This is a Normal Distribution where we notice that as the sample size increases the mean of the sample becomes closer to the expected value which is in this case my_mean = 27. This is a clear demonstration of the Law of Large numbers that states that the average of the results obtained from a large number of trials should be close to the expected value, and will become closer as more trials are performed.

## Poisson Distribution

```python
d['poisson'] = {}

for int in my_size:
    d['poisson'][int] = np.random.poisson(lam=my_mean, size=int)
    
len(d['poisson'].keys())
```

```python
d['poisson']['means'] = {}

fig2 = plt.figure(figsize=(12,6), dpi=100)
ax_poisson = fig2.add_subplot(111)
    
for smpl in d['poisson']:
    if not type(smpl) == str:
        d['poisson']['means'][smpl] = d['poisson'][smpl].mean()
        ax_poisson.plot(smpl, d['poisson']['means'][smpl], 'go')
        ax_poisson.set_title('Poisson Distribution', fontsize=20)
        ax_poisson.set_xlabel('Sample Size', fontsize=15)
        ax_poisson.set_ylabel('Sample Mean',fontsize=15)
        ax_poisson.plot([min(my_size), max(my_size)], [my_mean, my_mean], 'm--')
```

**Figure 2:** This is a Poisson Distribution where we notice that as the sample size increases the mean of the sample becomes closer to the expected value which is in this case my_mean = 27. This is a clear demonstration of the Law of Large numbers that states that the average of the results obtained from a large number of trials should be close to the expected value, and will become closer as more trials are performed.

## Chi-squared Distribution

**Figure 4:** This is a Chi-squared Distribution where we notice that as the sample size increases the mean of the sample becomes closer to the expected value which is in this case my_mean = 27. ## Laplace Distribution

```python
d['laplace'] = {}

for int in my_size:
    d['laplace'][int] = np.random.laplace(loc=my_mean, size=int)
    
len(d['laplace'].keys())
```

```python
d['laplace']['means'] = {}

fig5 = plt.figure(figsize=(12,6), dpi=100)
ax_laplace = fig5.add_subplot(111)
    
for smpl in d['laplace']:
    if not type(smpl) == str:
        d['laplace']['means'][smpl] = d['laplace'][smpl].mean()
        ax_laplace.plot(smpl, d['laplace']['means'][smpl], 'go')
        ax_laplace.set_title('Laplace Distribution', fontsize=20)
        ax_laplace.set_xlabel('Sample Size', fontsize=15)
        ax_laplace.set_ylabel('Sample Mean',fontsize=15)
        ax_laplace.plot([min(my_size), max(my_size)], [my_mean, my_mean], 'm--')
        
```

**Figure 5:** This is a Laplace Distribution where we notice that as the sample size increases the mean of the sample becomes closer to the expected value which is in this case my_mean = 27. This is a clear demonstration of the Law of Large numbers that states that the average of the results obtained from a large number of trials should be close to the expected value, and will become closer as more trials are performed.
